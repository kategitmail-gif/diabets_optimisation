```python
# PCA on crabs dataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.linalg import eigh
from numpy.linalg import svd

# 1) Load data
df = pd.read_csv("spike_data.csv")
crabs = pd.read_csv(url)
# Selected numeric columns in the exam screenshot
X = crabs[['FL', 'RW', 'CL', 'CW', 'BD']].to_numpy()  # n x p

n, p = X.shape
print("n, p =", n, p)

# Center the data (important for PCA)
mean_X = X.mean(axis=0)
Xc = X - mean_X

# (a) Sample covariance S (unbiased: divide by n-1)
S = np.cov(X, rowvar=False, bias=False)  # p x p
print("\n(a) Sample covariance S:\n", S)

# (b) Eigendecomposition of S
# Use eigh (for symmetric) which returns ascending eigenvalues
eigvals, eigvecs = eigh(S)
# sort descending
idx = np.argsort(eigvals)[::-1]
eigvals = eigvals[idx]
eigvecs = eigvecs[:, idx]
print("\n(b) Eigenvalues (descending):\n", eigvals)
print("\n(b) Principal components (eigenvectors, columns):\n", eigvecs)

# Check orthonormality
print("\nEigenvectors orthonormal? (V^T V):\n", eigvecs.T @ eigvecs)

# (c) PCA projections (scores): Xc @ eigenvectors
scores = Xc @ eigvecs  # n x p
# Usually keep first k columns for plotting, say first 3
df_scores = pd.DataFrame(scores[:, :3], columns=['PC1','PC2','PC3'])
# Attach species/class if you want to color:
if 'sp' in crabs.columns:
    hue = crabs['sp']
elif 'species' in crabs.columns:
    hue = crabs['species']
else:
    hue = None

print("\n(c) First 6 rows of PCA scores (PC1,PC2,PC3):\n", df_scores.head())

# Pairplot of the first two PCs (if seaborn available)
plt.figure(figsize=(6,6))
if hue is not None:
    sns.scatterplot(x=scores[:,0], y=scores[:,1], hue=hue, palette='deep')
else:
    plt.scatter(scores[:,0], scores[:,1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA: PC1 vs PC2')
plt.tight_layout()
plt.show()

# Optional full pairplot (can be slow):
# sns.pairplot(pd.concat([df_scores.iloc[:,:2], crabs.reset_index(drop=True)], axis=1),
#              vars=['PC1','PC2'], hue='sp' if hue is not None else None)

# (d) SVD decomposition of centered X
U, s, VT = svd(Xc, full_matrices=False)  # Xc = U * diag(s) * VT
print("\n(d) Singular values s:\n", s)
# Relationship: eigenvalues of S = (s^2)/(n-1)
eigvals_from_svd = (s**2) / (n - 1)
print("\nEigenvalues from SVD (s^2/(n-1)):\n", eigvals_from_svd)

# Principal directions from SVD: columns of VT.T (equal to eigvecs up to sign)
svd_pc = VT.T
# Align signs (make comparable)
for j in range(p):
    if np.allclose(np.abs(np.dot(svd_pc[:,j], eigvecs[:,j])), 0):
        # pass
        continue

# show max absolute differences column-wise (up to sign)
diffs = np.abs(np.abs(eigvecs) - np.abs(svd_pc))
print("\nMax abs difference between eigvecs and SVD PCs (abs, col-wise):", diffs.max(axis=0))

# (e) Gram matrix B = Xc Xc^T / (n-1)  (n x n)
B = (Xc @ Xc.T) / (n - 1)
print("\n(e) Gram matrix B shape:", B.shape)

# (f) Eigendecomposition of B -> obtain principal coordinates
# B u_i = lambda_i u_i, where lambda_i are same non-zero eigenvalues as S (but repeated with zeros)
eigvals_B, eigvecs_B = eigh(B)
# sort descending
idxb = np.argsort(eigvals_B)[::-1]
eigvals_B = eigvals_B[idxb]
eigvecs_B = eigvecs_B[:, idxb]

# Principal coordinates from B: for nonzero eigenvalues,
# coordinates = u_i * sqrt(lambda_i), i.e. columns = eigvecs_B[:,i] * sqrt(eigvals_B[i])
# Alternatively, rows of (U * s / sqrt(n-1)) match scores above (up to sign).
nonzero = eigvals_B > 1e-12
coords_from_B = eigvecs_B[:, nonzero] * np.sqrt(eigvals_B[nonzero])

print("\n(f) First 6 rows of coordinates from B (first 3 dims):\n", coords_from_B[:6,:3])

# Compare with 'scores' (which are Xc @ eigvecs). They should match (up to sign) for the first r components.
r = min(coords_from_B.shape[1], scores.shape[1])
# take first r
cmp = coords_from_B[:, :r] - scores[:, :r]
print("\nMax absolute difference between coords_from_B and scores (col-wise):", np.max(np.abs(cmp)))
print("\nIf differences are ~0 (up to sign), methods are consistent.")
```
